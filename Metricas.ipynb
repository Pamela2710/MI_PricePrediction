```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear una lista para almacenar los DataFrames
dfs_lstm = []

# Iterar sobre las versiones y tipos de conjunto de datos
versiones = [1, 3, 5, 7]
tipos = ['Tr']

for v in versiones:
    for t in tipos:
        # Cargar el DataFrame correspondiente
        df = pd.read_csv(project_path + f'metricas_L_Vt{v}/empresasLSTM{t}.csv', index_col=0)
        # Agregar una columna para indicar la versión y tipo de conjunto de datos
        df['Version'] = f'V{v}_{t}'
        # Agregar el DataFrame a la lista
        dfs_lstm.append(df)

# Concatenar todos los DataFrames en uno solo
df_concat = pd.concat(dfs_lstm)

# Plotear los datos
plt.figure(figsize=(12, 20))

# Iterar sobre las métricas
for i, metrica in enumerate(df_concat.columns[:-1]):  # Iterar sobre todas las columnas excepto 'Version'
    # Crear un subplot para cada métrica
    plt.subplot(len(df_concat.columns[:-1]), 1, i + 1)

    # Iterar sobre las versiones y tipos de conjunto de datos
    for v in versiones:
        for t in tipos:
            # Filtrar el DataFrame por versión y tipo de conjunto de datos
            df_temp = df_concat[df_concat['Version'] == f'V{v}_{t}']
            # Graficar los valores de la métrica para la versión y tipo de conjunto de datos actual
            plt.bar(df_temp.index, df_temp[metrica], label=f'V{v}_{t}')

    # Configuraciones del gráfico
    plt.title(f'Comparación de {metrica}')
    plt.xlabel('Empresa')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    plt.legend(title='Versión y Tipo de Conjunto de Datos', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
def get_company_split(df,metric):
    data=[[],[],[],[],[],[]]
    count=0
    for value in df[metric]:
        mod=count%6
        count+=1
        data[mod].append(value)
    return data

# def box_gen(df):
#     metrics=['MSE','MAE', 'MAPE', 'RMSE', 'MPE']
#     plt.figure(figsize=(10, 20))
#     for i in range(len(metrics)):
#         data=get_company_split(df,metrics[i])

#         plt.subplot(len(metrics),1,i+1)
#         plt.boxplot(data, labels=classes,vert=False)
#         plt.xlabel('Values')
#         plt.ylabel(f'{metrics[i]} Values')
#         plt.title('Companies')
#     plt.subplots_adjust(hspace=1)  # Adjust spacing between subplots
#     plt.show()

```

```python
def dataframe_por_empresa(df, classes):
    metrics = ['MSE', 'MAE', 'MAPE', 'RMSE', 'MPE']
    data_by_empresa = {}

    for metric in metrics:
        data_by_empresa[metric] = get_company_split(df, metric)

    return pd.DataFrame(data_by_empresa, index=classes)
```

```python
classes=['AAPL', 'AMZN', 'GOOG', 'GOOGL', 'MSFT', 'TSLA']
```

```python
# Función para calcular el promedio de una lista
def calcular_promedio(lista):
    if not lista:
        return 0  # Return 0 if the list is empty
    else:
        return sum(lista) / len(lista)
```
